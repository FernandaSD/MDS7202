En el siguiente proyecto se analizará un problema de detección de fraude en cuentas bancarias. Este es un dataset basado en datos reales que está compuesto por registros de aplicaciones para crear cuentas bancarias, entre las cuales se encuentra un bajo porcentaje de postulaciones fraudulentas. Se trata de un problema de aprendizaje no supervisado, a través del cual se buscarán anomalías en las aplicaciones a través de variables asociadas al postulante y al proceso de postulación. Entre las variables relacionadas al postulante se encuentra su ocupación, edad (redondeada a la decena), ingreso anual (en deciles), y entre las variables asociadas a la postulación se tiene el medio a través del cual se realiza la aplicación, la cantidad de días desde que esta fue realizada y otras variables de interés. 

Para la resolución del problema se plantea la hipótesis de que las postulaciones fraudulentas tendrán irregularidades que harán posibles su detección a través de modelos de aprendizaje no supervisado. Para realizar esto se propone hacer una exploración de datos para seleccionar las variables más relevantes, realizar un preprocesamiento de datos para dejar las variables en el formato y escala óptima, para luego realizar los clusters para detectar fraude. 

Para abordar el problema se seleccionó el modelo One Class Support Vector Machines. A diferencia de los SVM tradicionales que se utilizan para la clasificación, el One-Class SVM se entrena con un solo conjunto de datos que representa la clase normal, sin tener en cuenta los ejemplos de anomalías.

El objetivo del One-Class SVM es encontrar una frontera que englobe la región que contiene la mayoría de los datos normales, tratando de minimizar los datos que quedan fuera de esa región. Luego, se puede utilizar esta frontera para identificar instancias que caen fuera de ella como anomalías.

El modelo One-Class SVM es útil en escenarios en los que no se tienen ejemplos de anomalías etiquetadas, ya que solo se necesita información sobre la clase normal para entrenar el modelo.

Si bien este dataset sí contiene las etiquetas que indican si una postulación es fraudulenta o no (disponible en la variable ""fraud_bool), se trató el problema como un modelo de detección de anomalías y no como un problema de clasificación, usándose "fraud_bool" únicamente para evaluar el desempeño del modelo empleado a través de métricas como precision y recall. Cabe destacar que dado que se trata de un problema con un fuerte desbalance, es preferible asegurarse de que las postulaciones con fraude sean bien detectadas, dado que métricas como accuracy inducirán a que las postulaciones sin fraude mejoren este índice de desempeño, sin aportar realmente al problema.

```python
# Función de segmentación
def segmentar(age):
  if age < 18:
    return "Joven"
  elif age < 26:
    return "Adulto-Joven"
  elif age < 60:
    return "Adulto"
  else:
    return "Persona Mayor"
```

## 5. Preprocesamiento

En esta sección se aplicarán preprocesamientos sobre las columnas del dataset. En particular, se aplican las siguientes transformaciones:

- Variables categóricas: se aplicará One Hot Encoder.
- Variables ordinales: se aplicará Ordinal Encoder. El dataset contienen una única vaiable ordinal que es 'segmentacion_etaria'.
- Variables numéricas: Se aplicará MinMax Scaler dado que este escalador no afecta la escarcidad de los datos por lo que conserva la forma de la distribución en un rango menor (entre cero y uno). Se espera que esto facilite la detección de anomalías. 
- Variables binarias: No necesitan ser transformadas.
- Sobre las columnas 'id' y 'fraud_bool' se aplica 'passthrough' ya que no necesitan ser transformadas.

Antes de realizar esto, se trabajará con un dataframe sin las variables indicadas en la sección anterior, y sin los datos negativos (valores faltantes) que estaban presentes en algunas columnas. 

```python
# Eliminación de columnas
df_procesado = df_final.drop(columns = ['prev_address_months_count', 
                                        'customer_age', 'intended_balcon_amount',
                                        'x1', 'x2', 'device_fraud_count', 
                                        'device_distinct_emails_8w',
                                        'month', 'velocity_24h',
                                        'bank_branch_count_8w',
                                        'credit_risk_score']) 
```

```python
# Eliminación de datos faltantes
df_procesado = df_procesado[df_procesado['current_address_months_count'] >= 0]
df_procesado = df_procesado[df_procesado['bank_months_count'] >= 0]
df_procesado = df_procesado[df_procesado['session_length_in_minutes'] >= 0]
```

### 5.1 Declarar `ColumnTransformer`

Para aplicar las transformaciones primero se declara un 'Column Transformer' que contendrá todos los preprocesamientos que se aplicarán a las columnas correspondientes.

```python
df_procesado.info()
```

```python
# Se define las columnas y sus preprocesadores 

# El valor de id y fraud_bool no necesita ser transformado
id_col = ['id']
target_col = ['fraud_bool']

# Columnas categóricas
cat_cols = ['payment_type', 'employment_status', 'housing_status', 'source',
            'device_os']  
cat_transformer = OneHotEncoder(handle_unknown='ignore', sparse=False)

# Columnas ordinales
ord_cols = ['segmentacion_etaria']  
ord_transformer = OrdinalEncoder()

# Columnas numéricas
num_cols = ['income', 'current_address_months_count', 'days_since_request',
            'zip_count_4w', 'velocity_6h', 'velocity_4w', 'bank_months_count',
            'proposed_credit_limit', 'session_length_in_minutes', 
            'name_email_similarity', 'date_of_birth_distinct_emails_4w']  
num_transformer = MinMaxScaler()

# ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[('cat', cat_transformer, cat_cols),
        ('ord', ord_transformer, ord_cols),
        ('id', 'passthrough', id_col),
        ('target', 'passthrough', target_col)],
    remainder='passthrough', 
    sparse_threshold = 0)
```

La salida de la transformación debe ser un pandas dataframe, por lo que se utiliza 'set_output' sobre el 'Column Transformer'.

```python
# Salida en pandas
preprocessor = preprocessor.set_output(transform="pandas")
```

### 5.2 Transformar datos

Una vez definidas las transformaciones se utiliza 'Pipeline' para aplicarlas sobre todo el dataframe.

```python
# Pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor)])

# Se aplican las transformaciones
df_procesado = pipeline.fit_transform(df_procesado)
```

### 6.2 Proyectar y agregar proyecciones al dataframe de muestreo

Luego de tener la muestra, se crea una función que, dependiendo del método que reciba ('PCA' o 'UMAP'), realiza la proyección correspondiente sobre el dataset entregado y agrega estas a dicho dataframe. 

```python
# Semilla
random.seed(5517)

# Función para proyectar y agregar proyecciones
def projection(data, method):
    # Realizar la proyección
    if method == 'pca':
        projection = PCA(n_components=2, random_state = 20295).fit_transform(data)

        # Agregar proyecciones
        data['proyeccion0'] = projection[:, 0]
        data['proyeccion1'] = projection[:, 1]

    # Realizar la proyección
    elif method == 'umap':
        projection = UMAP(n_components=2, random_state = 20295).fit_transform(data)

        # Agregar proyecciones
        data['proyeccion0'] = projection[:, 0]
        data['proyeccion1'] = projection[:, 1]

    else:
        raise ValueError("Método de proyección debe ser 'pca' o 'umap'.")

    return data
```

Una vez hecha la función, se aplica sobre los datos de la muestra utilizando los dos métodos y se vizualizan los primeros 10 registros de los dataframes resultantes.

### 7.4 Calcular ratios y responder

Por último, se calculan dos ratios para evaluar los resultados obtenidos.

- P: Cantidad de ejemplos predichos correctamente como fraude / cantidad total de datos predichos como fraude. \\

- R: Cantidad de ejemplos predichos correctamente como fraude / cantidad total de ejemplos que eran realmente fraude.

```python
segmentos = [segmento_etario_1, segmento_etario_2, segmento_etario_3, segmento_etario_4]
```

```python
k=0
# Para cada segmento etario
for s in segmentos:
  k+=1

  # Se calcula la cantidad de ejemplos predichos correctamente como fraude
  n_true_positive = s[(s['fraud_bool_prediction']== 1) & (s['fraud_bool'] == 1)].shape[0]

  # Se calcula la cantidad total de ejemplos predichos como fraude
  n_fraude_pred = s['fraud_bool_prediction'].value_counts()[1]

  # Se calcula la cantidad total de ejemplos que realmente eran fraude
  n_fraude = s['fraud_bool'].value_counts()[1]

  # Se calculan e imprimen los ratios P y R
  print(f'segmento {k}')
  print('precision =', round((n_true_positive/n_fraude_pred)*100,3), '%')
  print('recall =', round((n_true_positive/n_fraude)*100,3), '%', '\n')
  print()
```

El indicador P es precisión y el indicador R es recall. 

La precisión se refiere a la proporción de transacciones clasificadas correctamente como fraude en comparación con el total de transacciones clasificadas como fraude. Es decir, mide la exactitud de las predicciones positivas del modelo. Una alta precisión significa que hay menos falsos positivos, es decir, menos transacciones legítimas clasificadas erróneamente como fraudes.

El recall, por otro lado, se refiere a la proporción de transacciones fraudulentas que se identifican correctamente en relación con el total de transacciones fraudulentas presentes en los datos. Mide la capacidad del modelo para encontrar todos los casos positivos. Un alto recall indica que hay menos falsos negativos, es decir, menos transacciones fraudulentas que se pasan por alto o se clasifican incorrectamente como transacciones legítimas.



En el contexto de la detección de fraudes, es importante considerar tanto la precisión como el recall, pero su relevancia puede variar según las circunstancias. Si el costo de los falsos positivos (clasificar incorrectamente una transacción legítima como fraude) es alto, se debe priorizar la precisión para minimizar estos casos. Por otro lado, si el costo de los falsos negativos (no detectar una transacción fraudulenta) es alto, se debe priorizar el recall para asegurarse de que la mayoría de los casos de fraude sean identificados. 

Para este caso, se tiene que el mejor desempeño se encontró para el segmento 1 con precisión de 1.014% y recall de 57.798%. Finalmente, se encuentra el segmento 3 con el peor desempeño que tuvo ambas métricas iguales a cero. Se destaca que pese a que se alcanzó un recall de 0,57 aprox para el segmento 1 y un 0,51 para el segmento 4, en ningún segmeno se logró superar una precisión superior al 1% lo cual indica que el modelo tiene un desempeño muy bajo.
